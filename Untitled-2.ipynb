```python
X = df.drop(['PRICE'],axis=1)
Y = df['PRICE']
X_train , X_test , Y_train , Y_test = train_test_split(X,Y, random_state=26 , test_size=0.3)
mlr = LinearRegression()
mlr.fit(X_train,Y_train)
```

```python
mlr.intercept_
```

```python
y_pred = mlr.predict(X_train)
r2 = metrics.r2_score(Y_train, y_pred)
print("R²:", r2)
mae = metrics.mean_absolute_error(Y_train, y_pred)
print("MAE:", mae)
rmse = np.sqrt(metrics.mean_squared_error(Y_train, y_pred))
print("RMSE:", rmse)
```

```python
df_compare = pd.DataFrame({
    "Actual": Y_train,
    "Predicted": y_pred
})
sns.scatterplot(x="Actual", y="Predicted", data=df_compare)

```

```python
sns.regplot(x="Actual", y="Predicted", data=df_compare, line_kws={"color":"red"})
```

```python
from numpy import poly
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly  = poly.transform(X_test)

# مدل رگرسیون خطی روی داده‌های چندجمله‌ای
model_poly = LinearRegression()
model_poly.fit(X_train_poly, Y_train)

# پیش‌بینی
y_train_pred = model_poly.predict(X_train_poly)
y_test_pred  = model_poly.predict(X_test_poly)
```

```python
df_compare_poly = pd.DataFrame({
    "Actual": Y_train,
    "Predicted": y_train_pred
})
sns.regplot(x="Actual", y="Predicted", data=df_compare_poly, line_kws={"color":"red"})

```
